```
import graphviz
def drawUF(a):
    dot = graphviz.Digraph(comment='Nada')
    n = len(a)
    for i in range(n):
        dot.node(str(i), str(i))
    for i in range(n):
        if i != a[i]:
            dot.edge(str(i), str(a[i]))
    dot.graph_attr['rankdir'] = 'BT'
    return dot
```

# Union Find

## Quick Union

```
def QU_Find(id, a):
    while id[a] != a:
        a = id[a]

    return a

def quickUnion(id, a, b):
    parentA = QU_Find(id, a)
    parentB = QU_Find(id, b)
    if parentA == parentB:
        return
    id[parentA] = parentB
```

```
id = [i for i in range(10)]; drawUF(id)
```

```
quickUnion(id, 3, 4); print(id); drawUF(id)
```
